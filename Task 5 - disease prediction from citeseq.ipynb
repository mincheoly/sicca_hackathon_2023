# Task 5 - cell type proportions

```python
import scanpy as sc
import muon as mu
import pandas as pd
import itertools
pd.set_option('display.max_columns', 100)
```

```python
mdata = mu.read('task5_input.h5mu')
```

```python
prot_expr = pd.DataFrame(
    mdata.mod['prot'].X.toarray(),
    columns=mdata.mod['prot'].var.index)
prot_expr['siccaid'] = mdata.obs['siccaid'].tolist()
prot_expr['ct'] = mdata.obs['annotation_merged'].tolist()
```

```python
avg_prot_expr = prot_expr.groupby(['siccaid', 'ct']).mean()
```

```python
ind_list = mdata.obs.siccaid.drop_duplicates().tolist()
prot_list = mdata.mod['prot'].var.index.tolist()
ct_list = (mdata.obs['annotation_merged'].value_counts() > 1000).index.tolist()#mdata.obs.annotation_merged.drop_duplicates().tolist()
```

```python
num_features = len(ct_list) * len(prot_list)
num_obs = len(ind_list)

X = pd.DataFrame(index=ind_list, columns=[x + '_' + y for x,y in itertools.product(ct_list, prot_list)])
```

```python
for ind, ct, prot in itertools.product(ind_list, ct_list, prot_list):

    try:
        X.loc[ind, ct + '_' + prot] = avg_prot_expr.loc[(ind, ct), prot]
    except:
        continue
X = X.fillna(0.0)
```

```python
y = mdata.obs[['siccaid', 'fs', 'pSS', 'age', 'sex', 'fs_detailed', 'lympho', 'rna:ssa', 'rna:ssb']].drop_duplicates().set_index('siccaid').clip(lower=0, upper=1)
```

```python
Xy = X.join(y)
```

```python
Xy.to_csv('prediction3.csv')
```
